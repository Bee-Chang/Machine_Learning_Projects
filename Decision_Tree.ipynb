```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import numpy as np
```

```python
df = pd.read_csv('fetal_health.csv')
df
```

```python
x_data=df[['baseline value', 'accelerations', 'fetal_movement',
       'uterine_contractions', 'light_decelerations',
       'severe_decelerations', 'prolongued_decelerations',
       'abnormal_short_term_variability',
       'mean_value_of_short_term_variability',
       'percentage_of_time_with_abnormal_long_term_variability',
       'mean_value_of_long_term_variability', 'histogram_width',
       'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
       'histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean',
       'histogram_median', 'histogram_variance', 'histogram_tendency']]
y_data=df['fetal_health']
```

```python
df.columns.values
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=99) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

Drug Classification Dataset

```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import numpy as np
```

```python
data = pd.read_csv('drug.csv')
data
```

```python
data = data.dropna()
data
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

label = le.fit_transform(data['BP']) 
label1 = le.fit_transform(data['Cholesterol']) 
label2 = le.fit_transform(data['Drug'])

label
label1
```

```python
data.drop("BP", axis=1, inplace=True)
data.drop("Cholesterol", axis=1, inplace=True)
data.drop("Drug", axis=1, inplace=True)

data["BP"] = label 
data["Cholesterol"] = label1
data["Drug"] = label2

data.head(15)
```

```python
x_data=data[['Age','Na_to_K','BP','Cholesterol']]
y_data=data['Drug']
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3, random_state=99) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import graphviz 
import numpy as np


# Load data from CSV file
new_cols = ['Age','Na_to_K','BP','Cholesterol', 'Drug']
data = data.reindex(columns=new_cols)

data.fillna(value=0, inplace=True)

# Split data into features and target
X = data.drop(columns=['Age','Na_to_K','BP','Cholesterol'])
y = data['Drug']

# Fit decision tree model
model = DecisionTreeClassifier()
model.fit(X, y)

# Generate decision tree visualization
dot_data = tree.export_graphviz(model, out_file=None, 
                     feature_names=X.columns,  
                     class_names=y.unique(),  
                     filled=True, rounded=True,  
                     special_characters=True)  

graph = graphviz.Source(dot_data)  
graph.render("decision_tree") 
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

import dtreeviz

iris = load_iris()
X = iris.data
y = iris.target

clf = DecisionTreeClassifier(max_depth=4)
clf.fit(X, y)

viz_model = dtreeviz.model(clf,
                           X_train=X, y_train=y,
                           feature_names=iris.feature_names,
                           target_name='iris',
                           class_names=iris.target_names)

v = viz_model.view()     # render as SVG into internal object 
v.show()                 # pop up window
v.save("/tmp/iris.svg")
```
