```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
```

```python
df = pd.read_csv('Stars.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

Label = le.fit_transform(df['Color'])

df.drop('Color', axis=1, inplace=True)

df['Color'] = Label

df
```

```python
Label1 = le.fit_transform(df['Spectral_Class'])

df.drop('Spectral_Class', axis=1, inplace=True)

df['Spectral_Class'] = Label1

df
```

```python
df.columns.values
```

```python
#the x data is saved without 'target' column to make only features dataframe
x=df[['Temperature', 'L', 'R', 'A_M', 'Color', 'Spectral_Class']]
```

```python
#the y data is saved only with 'target' column
y=df['Type']
```

```python
#create a model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
#kernel specifying
clf=svm.SVC(kernel='linear')

#train and fit the model
clf.fit(x_train,y_train)
```

```python
#predict the response
y_pred=clf.predict(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

Label2 = le.fit_transform(df1['microorganisms'])

df1.drop('microorganisms', axis=1, inplace=True)

df1['microorganisms'] = Label2

df1
```

```python
df1.columns.values
```

```python
#the x data is saved without 'target' column to make only features dataframe
x=df1[['Solidity', 'Eccentricity', 'EquivDiameter',
       'Extrema', 'FilledArea', 'Extent', 'Orientation', 'EulerNumber',
       'BoundingBox1', 'BoundingBox2', 'BoundingBox3', 'BoundingBox4',
       'ConvexHull1', 'ConvexHull2', 'ConvexHull3', 'ConvexHull4',
       'MajorAxisLength', 'MinorAxisLength', 'Perimeter', 'ConvexArea',
       'Centroid1', 'Centroid2', 'Area', 'raddi']]
```

```python
y=df1['microorganisms']
```

```python
#create a model
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
#kernel specifying
clf=svm.SVC(kernel='linear')

#train and fit the model
clf.fit(x_train,y_train)
```

```python
#predict the response
y_pred=clf.predict(x_test)
```

```python
#model accuracy
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred))
```

```python
precision = metrics.precision_score(y_test, y_pred, average='macro')
recall = metrics.recall_score(y_test, y_pred, average='micro')

print("Precision:", precision)
print("Recall:", recall)
```
